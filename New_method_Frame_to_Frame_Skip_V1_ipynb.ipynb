```python
def convert_drive_url_to_downloadable(drive_url):
    # ดึงไฟล์ ID จากลิงก์ Google Drive
    file_id = drive_url.split('/d/')[1].split('/')[0]
    # แปลงเป็นลิงก์ดาวน์โหลด
    download_url = f'https://drive.google.com/uc?export=download&id={file_id}'
    return download_url

# สร้างตัวแปร df1 ถึง df50
dfs = {}

# อ่าน CSV ทีละ URL และเก็บในตัวแปร df1 ถึง df50
for i, url in enumerate(url_list):
    if i < 60:  # จำกัดที่ 50 แถว
        try:
            # แปลงลิงก์ Google Drive เป็นลิงก์ดาวน์โหลด
            download_url = convert_drive_url_to_downloadable(url)
            # อ่านไฟล์ CSV จากลิงก์ที่ถูกแปลงแล้ว
            dfs[f'df{i+1}'] = pd.read_csv(download_url)
        except Exception as e:
            print(f"Error reading CSV at URL {url}: {e}")
```

```python
dfss_odd = []
dfss_even = []
dfss_combined = []

for i in range(len(dfss)):
    df = dfss[i]

    # แยกแถวคี่และคู่ในแต่ละ DataFrame
    odd_rows = df.iloc[1::2]  # เลือกแถวคี่ (index เริ่มที่ 1, ข้ามทีละ 2)
    even_rows = df.iloc[0::2]  # เลือกแถวคู่ (index เริ่มที่ 0, ข้ามทีละ 2)

    # รวมแถวคี่ทั้งหมดเป็น 1 row
    flattened_odd = pd.concat([odd_rows.iloc[j] for j in range(len(odd_rows))]).reset_index(drop=True)

    # รวมแถวคู่ทั้งหมดเป็น 1 row
    flattened_even = pd.concat([even_rows.iloc[j] for j in range(len(even_rows))]).reset_index(drop=True)

    flattened_odd = flattened_odd.iloc[:106350]
    flattened_even = flattened_even.iloc[:106350]

    combined_row = pd.concat([flattened_even, flattened_odd], axis=1).reset_index(drop=True)

    # แปลงกลับเป็น DataFrame ที่มี 1 แถว
    dfss_odd.append(pd.DataFrame([flattened_odd]))
    dfss_even.append(pd.DataFrame([flattened_even]))
    dfss_combined.append(pd.DataFrame([combined_row.values.flatten()]))
```

```python
combined_df = pd.concat(dfss_even, axis=0).reset_index(drop=True) #เลือกแค่ even รันเฟมที่ 1,3,5 ไปเรื่อยๆ
```

```python
train_df = full[full['random_train/test'] == 'Train']
test_df = full[full['random_train/test'] == 'Test']
```

```python
X_train = train_df.iloc[:,2:]
X_test = test_df.iloc[:,2:]
```

```python
y_train = train_df.iloc[:,0:1]
y_test = test_df.iloc[:,0:1]
```

Classification

RF

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")

if train_accuracy > test_accuracy:
    print("โมเดล Overfit")
else:
    print(" ไม่มีปัญหา Overfitting")
```

SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)

# Evaluate
print(classification_report(y_test, y_pred))
```

```python
y_train_predsvm = svm.predict(X_train)
y_test_predsvm = svm.predict(X_test)

train_accuracysvm = accuracy_score(y_train, y_train_predsvm)
test_accuracysvm = accuracy_score(y_test, y_test_predsvm)

print(f"Train Accuracy: {train_accuracysvm}")
print(f"Test Accuracy: {test_accuracysvm}")

if train_accuracysvm > test_accuracysvm:
    print("โมเดล Overfit")
else:
    print(" ไม่มีปัญหา Overfitting")
```
