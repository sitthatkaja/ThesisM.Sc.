```python
def convert_drive_url_to_downloadable(drive_url):
    # ดึงไฟล์ ID จากลิงก์ Google Drive
    file_id = drive_url.split('/d/')[1].split('/')[0]
    # แปลงเป็นลิงก์ดาวน์โหลด
    download_url = f'https://drive.google.com/uc?export=download&id={file_id}'
    return download_url

# สร้างตัวแปร df1 ถึง df50
dfs = {}

# อ่าน CSV ทีละ URL และเก็บในตัวแปร df1 ถึง df50
for i, url in enumerate(url_list):
    if i < 60:  # จำกัดที่ 50 แถว
        try:
            # แปลงลิงก์ Google Drive เป็นลิงก์ดาวน์โหลด
            download_url = convert_drive_url_to_downloadable(url)
            # อ่านไฟล์ CSV จากลิงก์ที่ถูกแปลงแล้ว
            dfs[f'df{i+1}'] = pd.read_csv(download_url)
        except Exception as e:
            print(f"Error reading CSV at URL {url}: {e}")
```

```python
dfss_odd = []
dfss_even = []
dfss_combined = []

for i in range(len(dfss)):
    df = dfss[i]

    # แยกแถวคี่และคู่ในแต่ละ DataFrame
    odd_rows = df.iloc[1::2]  # เลือกแถวคี่ (index เริ่มที่ 1, ข้ามทีละ 2)
    even_rows = df.iloc[0::2]  # เลือกแถวคู่ (index เริ่มที่ 0, ข้ามทีละ 2)

    # รวมแถวคี่ทั้งหมดเป็น 1 row
    flattened_odd = pd.concat([odd_rows.iloc[j] for j in range(len(odd_rows))]).reset_index(drop=True)

    # รวมแถวคู่ทั้งหมดเป็น 1 row
    flattened_even = pd.concat([even_rows.iloc[j] for j in range(len(even_rows))]).reset_index(drop=True)

    flattened_odd = flattened_odd.iloc[:106350]
    flattened_even = flattened_even.iloc[:106350]

    combined_row = pd.concat([flattened_even, flattened_odd], axis=1).reset_index(drop=True)

    # แปลงกลับเป็น DataFrame ที่มี 1 แถว
    dfss_odd.append(pd.DataFrame([flattened_odd]))
    dfss_even.append(pd.DataFrame([flattened_even]))
    dfss_combined.append(pd.DataFrame([combined_row.values.flatten()]))
```

RF

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
loo = LeaveOneOut()
TP = 0
TN = 0
FP = 0
FN = 0
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # แปลง y_train ให้อยู่ในรูปแบบ 1D
    y_train = y_train.ravel()

    # ฝึกโมเดล
    clf.fit(X_train, y_train)

    # ทำนายผล
    y_pred = clf.predict(X_test)[0]
    y_true = y_test[0]

    # เก็บค่าผลลัพธ์ TP, TN, FP, FN
    if y_true == 1 and y_pred == 1:
        TP += 1
    elif y_true == 0 and y_pred == 0:
        TN += 1
    elif y_true == 0 and y_pred == 1:
        FP += 1
    elif y_true == 1 and y_pred == 0:
        FN += 1

# คำนวณเมตริกจาก TP, TN, FP, FN
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP) if (TP + FP) > 0 else 0
recall = TP / (TP + FN) if (TP + FN) > 0 else 0
f1_score = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

# แสดงผลเมตริก
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1_score}")

# แสดงผลค่า TP, TN, FP, FN
print("\nConfusion Matrix Counts:")
print(f"TP: {TP}, TN: {TN}, FP: {FP}, FN: {FN}")
```
