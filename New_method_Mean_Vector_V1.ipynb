```python
df = pd.DataFrame(worksheet.get_all_records())
df
```

```python
status = df.iloc[:,2:4]
status
```

```python
full = pd.concat([status,result], axis=1)
full
```

```python
train_df = full[full['random_train/test'] == 'Train']
test_df = full[full['random_train/test'] == 'Test']
```

```python
X_train = train_df.iloc[:,2:]
X_test = test_df.iloc[:,2:]
```

```python
y_train = train_df.iloc[:,0:1]
y_test = test_df.iloc[:,0:1]
```

Classification

RF

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")

if train_accuracy > test_accuracy:
    print(" Overfit")
else:
    print("  Overfitting")
```

SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)

# Evaluate
print(classification_report(y_test, y_pred))
```

```python
y_train_predsvm = svm.predict(X_train)
y_test_predsvm = svm.predict(X_test)

train_accuracysvm = accuracy_score(y_train, y_train_predsvm)
test_accuracysvm = accuracy_score(y_test, y_test_predsvm)

print(f"Train Accuracy: {train_accuracysvm}")
print(f"Test Accuracy: {test_accuracysvm}")

if train_accuracysvm > test_accuracysvm:
    print(" Overfit")
else:
    print("  Overfitting")
```
