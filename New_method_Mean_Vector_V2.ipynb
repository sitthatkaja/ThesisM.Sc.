```python

def convert_drive_url_to_downloadable(drive_url):
    # ดึงไฟล์ ID จากลิงก์ Google Drive
    file_id = drive_url.split('/d/')[1].split('/')[0]
    # แปลงเป็นลิงก์ดาวน์โหลด
    download_url = f'https://drive.google.com/uc?export=download&id={file_id}'
    return download_url

# สร้างตัวแปร df1 ถึง df50
dfs = {}

# อ่าน CSV ทีละ URL และเก็บในตัวแปร df1 ถึง df50
for i, url in enumerate(url_list):
    if i < 50:  # จำกัดที่ 50 แถว
        try:
            # แปลงลิงก์ Google Drive เป็นลิงก์ดาวน์โหลด
            download_url = convert_drive_url_to_downloadable(url)
            # อ่านไฟล์ CSV จากลิงก์ที่ถูกแปลงแล้ว
            dfs[f'df{i+1}'] = pd.read_csv(download_url)
        except Exception as e:
            print(f"Error reading CSV at URL {url}: {e}")
```

```python
# สร้าง DataFrame ว่างเพื่อเก็บผลลัพธ์
mean_df = pd.DataFrame()

# ลูปผ่านแต่ละ DataFrame และคำนวณค่าเฉลี่ย
for i, df in enumerate(dfss, start=1):
    mean_series = df.mean()  # คำนวณค่าเฉลี่ยของแต่ละคอลัมน์ใน DataFrame
    mean_df[f'df{i}_mean'] = mean_series  # เพิ่มแถวใหม่ใน mean_df เพื่อเก็บค่าเฉลี่ย

transposed_mean_df = mean_df.T
transposed_mean_df
```

```python
full = pd.concat([status,result], axis=1)
full
```

```python
train_df = full[full['random_train/test'] == 'Train']
test_df = full[full['random_train/test'] == 'Test']
```

```python
X_train = train_df.iloc[:,2:]
X_test = test_df.iloc[:,2:]
```

```python
y_train = train_df.iloc[:,0:1]
y_test = test_df.iloc[:,0:1]
```

Classification

RF

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")

if train_accuracy > test_accuracy:
    print("โมเดล Overfit")
else:
    print(" ไม่มีปัญหา Overfitting")
```
