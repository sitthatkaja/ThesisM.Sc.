```python
import numpy as np
```

RF

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
loo = LeaveOneOut()
accuracies = []
precisions = []
recalls = []
f1_scores = []
predictions = []  # เก็บค่าการทำนาย
actuals = []  # เก็บค่าจริง
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
for train_index, test_index in loo.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # ฝึกโมเดล
    clf.fit(X_train, y_train)

    # ทำนายผล
    y_pred = clf.predict(X_test)

    # เก็บผลการทำนาย
    predictions.append(y_pred[0])
    actuals.append(y_test[0])

    # คำนวณเมตริก
    acc = accuracy_score([y_test[0]], y_pred)
    prec = precision_score([y_test[0]], y_pred, average='binary', zero_division=0) if len(set(y)) == 2 else precision_score([y_test[0]], y_pred, average='macro', zero_division=0)
    rec = recall_score([y_test[0]], y_pred, average='binary', zero_division=0) if len(set(y)) == 2 else recall_score([y_test[0]], y_pred, average='macro', zero_division=0)
    f1 = f1_score([y_test[0]], y_pred, average='binary', zero_division=0) if len(set(y)) == 2 else f1_score([y_test[0]], y_pred, average='macro', zero_division=0)

    accuracies.append(acc)
    precisions.append(prec)
    recalls.append(rec)
    f1_scores.append(f1)

# คำนวณค่าเฉลี่ยของเมตริกทั้งหมด
mean_accuracy = np.mean(accuracies)
mean_precision = np.mean(precisions)
mean_recall = np.mean(recalls)
mean_f1_score = np.mean(f1_scores)

print(f"Mean Accuracy from LOOCV: {mean_accuracy}")
print(f"Mean Precision from LOOCV: {mean_precision}")
print(f"Mean Recall from LOOCV: {mean_recall}")
print(f"Mean F1 Score from LOOCV: {mean_f1_score}")

# แสดงผลการทำนายและค่าจริง
print("\nPredictions vs Actuals:")
for i, (pred, actual) in enumerate(zip(predictions, actuals)):
    print(f"Sample {i + 1}: Predicted = {pred}, Actual = {actual}")
```
