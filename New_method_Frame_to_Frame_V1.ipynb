```python
import gspread
import gdown
from oauth2client.service_account import ServiceAccountCredentials
```

```python
df_links = pd.DataFrame(df)
for index, csv_url in enumerate(df_links['url_link_datacsv_link_Drive']):
    try:
        # ดึง ID ไฟล์จากลิงก์ Google Drive
        file_id = csv_url.split('/d/')[1].split('/')[0]
        download_url = f'https://drive.google.com/uc?id={file_id}'

        # ดาวน์โหลดไฟล์ CSV
        output_file = f'downloaded_file_{index+1}.csv'  # ตั้งชื่อไฟล์ให้ไม่ซ้ำกัน
        gdown.download(download_url, output_file, quiet=False)

        # โหลดไฟล์ CSV เข้าสู่ DataFrame
        df_csv = pd.read_csv(output_file)

        # สร้างตัวแปร DataFrame แบบไดนามิก (df1, df2, ...)
        exec(f"df{index+1} = pd.DataFrame(df_csv)")

        # แสดงข้อมูลของ DataFrame ที่โหลดได้
        print(f"ข้อมูลใน df{index+1}:")
        exec(f"print(df{index+1}.head())")

    except Exception as e:
        print(f"ไม่สามารถดาวน์โหลดหรือโหลดไฟล์จากลิงก์: {csv_url}")
        print(f"ข้อผิดพลาด: {e}")
```
